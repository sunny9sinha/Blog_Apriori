```python
# Import necessary libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Load "store_data.csv" dataset
store_data_df = pd.read_csv("/content/store_data.csv", header=None)

# Preprocess dataset
transactions = []
for i in range(len(store_data_df)):
    transactions.append([str(store_data_df.values[i,j]) for j in range(len(store_data_df.columns))])

# Transform dataset into one-hot encoding
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Generate frequent itemsets with minimum support of 0.01
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)

# Generate association rules with minimum confidence of 0.5
association_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```
